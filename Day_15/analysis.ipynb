```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
categories = pd.Series([25, 18, 32, 14, 27, 35, 21, 29, 19, 23],
                       index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
```

```python
sum = categories.sum()
print("Sum:", sum)
# to calculate the total profit
```

```python
mean= categories.mean()
median= categories.median()
std= categories.std()
mode = categories.mode()[0]

print("Mean:", mean)
print("Median:", median)
print("Standard deviation:", f"{std:.1f}")
print("Mode:",mode)

#  دايما بيتحسبوا وليهم استخدام مش عارفة ايه بقى
```

```python
max = categories.max()
min = categories.min()
count = categories.count()

print("Max:", max)
print("Min:", min)
print("Count:", count)

# to see which category made the highest sales and which one made the lowest
```

```python
categories.plot(kind='bar', color='lightpink')
plt.title('Chart of Categories and Values')
plt.xlabel('Category')
plt.ylabel('Values')

plt.axhline(y=mean, color='lightblue', linestyle='-',label=f'Mean: {mean}')
plt.axhline(y=median, color='lightgreen', linestyle='-', label=f'Median: {median}')
plt.axhline(y=std, color='lightcoral', linestyle='-', label="std: "f"{std:.1f}")
plt.axhline(y=mode, color='gray', linestyle='-', label=f'Mode: {mode}')
plt.axhline(y=min, color='purple', linestyle='-', label=f'Min: {min}')
plt.axhline(y=max, color='brown', linestyle='-', label=f'Max: {max}')

plt.legend()
plt.show()
```
